# Importing python libraries

```python
import pandas as pd
import numpy as np
```

# Loading dataset

```python
data=pd.read_csv("/kaggle/input/full-filled-brain-stroke-dataset/full_filled_stroke_data (1).csv")
```

```python
data
```

# Checking null Values 

```python
data.isnull().sum()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["gender"]=le.fit_transform(data["gender"])
data["ever_married"]=le.fit_transform(data["ever_married"])

data["work_type"]=le.fit_transform(data["work_type"])

data["smoking_status"]=le.fit_transform(data["smoking_status"])
data["Residence_type"]=le.fit_transform(data["Residence_type"])

data
```

```python
 data["stroke"].value_counts()
```

# Slicing dataset

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values

```

Imbalencing Technique 
# SMOTE

```python
from imblearn.over_sampling import SMOTE
s=SMOTE()
x_data,y_data=s.fit_resample(x,y)
```

```python

data.corr()
```

# Feature Scaling

```python

from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
x_scaled=mm.fit_transform(x_data)
```

# Spliting dataset in train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_data,test_size=0.2,random_state=11)
```

# Model  fitting

```python
from sklearn.svm import SVC
ss=SVC(kernel="rbf",random_state=3)
ss.fit(x_train,y_train)
```

```python
y_pred=ss.predict(x_test)
y_pred
```

# Accuracy Score

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)*100
ac
```
